```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

negara =('Argentina', 'Brazil', 'Chile', 'Denmark', 'Finlandia', 'German', 'Hongaria', 'Indonesia', 'Jepang', 'Korea Selatan')
populasi = (45380000, 212600000, 19120000, 5831000, 5531000, 83240000, 975000, 273500000, 12580000, 51780000)
kor_x = np.arange(len(negara))

df = pd.DataFrame({'Country':negara, 'Population':populasi})
df.sort_values(by='Population', inplace=True)

plt.bar(kor_x, df['Population'], tick_label=df['Country'])
plt.xticks(rotation = 90)
plt.ylabel("Populasi(Juta)")
plt.title('Populasi 10 Negara Di Dunia')
plt.show()
```
