# Soal 2 : Permen
*   Saya menggunakan Pie Chart karena pie char dapat menampilkan persentase dari data yang digunakan
*   Peluang memilih permen kopiko sebesar 15% atau 0,15

# Soal 3 : Makanan
*   Saya menggunakan Bar Chart karena dapat memvisualisasikan data penjualan dari penjualan terbanyak sampai penjualan terendah atau sebaliknya
*   Makanan penutup yang harus disingkirkan yaitu puding vanila, pastel, dan wajik



```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

datapenjualan_makananpenutup =  {
    'Donat': 14,  
    'Pastel': 5,  
    'Kue Coklat': 12,  
    'Ice Cream': 19,  
    'Puding Vanila': 6,  
    'Brownies': 8,  
    'Puding Strawberi': 12,  
    'Puding Coklat': 9,  
    'Ice Cream Nutela': 10, 
    'Kue Coklat-Keju': 17,  
    'Kue Wajik': 2,  
    'Kue Sus': 9,  
    'Mochi': 13, 
}

data_key = list(datapenjualan_makananpenutup.keys())
data_value = list(datapenjualan_makananpenutup.values())

df = pd.DataFrame({'Makanan':data_key, 'Penjualan':data_value})
df.sort_values(by='Penjualan', ascending = False, inplace = True)

kor_x = np.arange(len(df))
warna = ['#0000FF' for _ in range(len(df))]
warna [-1] = '#FF0000'
warna [-2] = '#FF0000'
warna [-3] = '#FF0000'

plt.figure(figsize=(20,10))

plt.bar(kor_x, df['Penjualan'], tick_label=df['Makanan'], color = warna)
plt.xticks(rotation = 90)
plt.ylabel("Data Penjualan")
plt.xlabel("Jenis Makanan Penutup")
plt.title('Data Penjualan Menu Makanan Penutup di Kafe Biru')
plt.show()
```
